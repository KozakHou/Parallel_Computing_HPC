```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
files = sorted(glob.glob('Result/output_*.dat'), key=lambda x: int(x.split('_')[-1].split('.')[0]))
files
```

```python
import matplotlib.animation as animation

# Create figure and subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('MHD Simulation Results')

# Initialize empty plots
line1, = ax1.plot([], [])
line2, = ax2.plot([], [])
line3, = ax3.plot([], [])
line4, = ax4.plot([], [])

ax1.set_title('Density')
ax2.set_title('Velocity')
ax3.set_title('Pressure')
ax4.set_title('Magnetic Field')

# Set axis limits
for ax in (ax1, ax2, ax3, ax4):
    ax.set_xlim(0, 1)  # Assuming x range is [0, 1]
    ax.set_ylim(0, 2)  # Adjust as needed

def animate(i):
    file = files[i]
    data = np.loadtxt(file)
    x = data[:, 0]
    rho = data[:, 1]
    u = data[:, 2]
    p = data[:, 3]
    B = data[:, 4]

    line1.set_data(x, rho)
    line2.set_data(x, u)
    line3.set_data(x, p)
    line4.set_data(x, B)

    fig.suptitle(f'Results at {file}')
    return line1, line2, line3, line4

anim = animation.FuncAnimation(fig, animate, frames=len(files), interval=200, blit=True)
anim.save('mhd_simulation.gif', writer='pillow')
plt.close()
```
